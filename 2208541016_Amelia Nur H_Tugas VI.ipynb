# 5.22.1 Grafik Sederhana dengan Matplotlib

```python
#contoh grafik sederhana
#misalkan dulu variabel x dan y merupakan list dari nomor yang berisi 4 nilai di dalam masing-masing variabel
x = [2,4,6,8]
y = [3,5,7,9]
#menggunakan kode plot yang akan menggambar garis yang menghubungkan keempat titik yaitu (2,3), (4,5), dan (6,7) dan (8,9) pada bidang koordinat
plt.plot (x,y)
#Kemudian, kode dibawah digunakan untuk memberikan judul grafik yaitu grafik y = x+2
plt.title ("Grafik y = x+2")
#sehingga, sumbu x diberi label dengan kode sebagai berikut
plt.xlabel ('x')
#sama juga dengan label pada sumbu y
plt.ylabel ('y')
```

# 5.22.2 Properti Pembuatan Grafik dengan Matplotlib

```python
#permisalkan variabel x dan y yang merupakan list yang berisi 5 nilai dari masing-masing variabel x dan y yang berkaitan dengan fungsi y = x^3
x = [-2,-1,0,1,2]
y = [-8,-1,0,1,8]
#lalu,permisalkan juga variabel x1 dan y1 merupakan list yang berisi masing-masing anggota dari variabel x1 dan y1 
x1 = [3,4,5,6,7]
y1 = [2,4,6,8,10]

plt.figure(figsize=(10,10), dpi =400)

plt.plot(x,y,label='y=x^3',color='yellow',linewidth=5,marker='.',markersize =10,markeredgecolor='green',linestyle='-')
#label='y=x^3' : dugunakan untuk memberi label atau judul pada plot

plt.plot(x1,y1)
#untuk membuat plot dari data pada variabel x1 dan y1
#plot ini tidak memiliki variabel lain seperti label ataupun color sehingga plot ini akan menggunakan nilai default untuk setiap parameter

plt.title ('Grafik y =x^3', fontdict={'fontname' : 'Times New Roman', 'fontsize' :18})
#Ditentukan font dan ukuran font pada judul grafik
#'Grafik y =x^3', adalah string yang akan ditampilkan sebagai judul grafik
#fontdict : digunakan untuk mengatur jenis font dan ukuran font dari judul grafik
#fontname : jenis fnt
#fontsize : ukuran font
plt.xlabel ('x', fontdict={'fontname':'Comic Sans MS','fontsize':16})
plt.ylabel ('y', fontdict={'fontname':'Comic Sans MS','fontsize':16})

#lalu,pada sumbu x dan y diberi label pada grafik
plt.xticks ([-2,-1,0,1,2])
plt.yticks ([-8,-1,0,1,8])
#Digunakan fungsi plt.ticks untuk menentukan skala sumbu x dan t yang akan ditampilkan pada grafik
#pada sumbu x ditampilkan angka dari -2 sampai 2
#pada sumbu y ditampilkan anhka dari -8 sampai 8
plt.legend()
#kode ini digunakan untuk menampilkan legenda pada grafik yang menunjukkan label garis untuk grafik pertaman
plt.savefig('grafik.png', dpi=400)
#kode digunakan untuk menyimpan grafik sebagai file gambar dalam format PNG dan nama Grafik "grafik.png" dengan resolusi 400dpi
plt.show()
#kode ini digunakan untuk menampilkan grafik hasilnya
```

# 5.22.4 Properti pembuaatan histogram dengan matplotlib

```python
#permisalkan sebuah list yang berisi beberapa kategori pekerjaan di sebuah desa
Pekerjaan = ['petani', 'wiraswasta','dokter', 'pedagang']
#lalu,permisalkan jumlah dari list data kategori pada setiap masing-masing pekerjaan
jumlah = [10,23,5,24]

plt.figure(figsize=(6,3))
#plt.figure untuk menentukan ukuran dan resolusi gambar
#figsize untuk menentukan gambar dengan format inci

histogram = plt.bar(Pekerjaan,jumlah,color = 'yellow')
#memanggil kode histogram dengan ditambahkan juga warna yang kita inginkan
pola = ['/','.','*','x']
#bisa juga memasukkan pola dalam sebuah list yang berisi simbol sesuai dengan keinginan kita yang ada pada variabel'histogram'
histogram [0].set_hatch('/')
histogram [1].set_hatch('.')
histogram [2].set_hatch('*')
histogram [3].set_hatch('x')
plt.show()
#lalu kita dapat memanggil kode di atas mejadi histogram
```

```python
#lalu, melakukan pembacaan data pada bunga_iris dengan kode read dari data dengan file yang sama seperti yang diatas tadi
data_iris = pd.read_csv('bunga_iris.csv')

#lalu,pilih kolom yang akan dimasukkan dalam gambar
petal_length = data_iris['petal length (cm)']
petal_width = data_iris['petal width (cm)']

#lalu,dibuat scatterplot,berkaitan dengan hubungan petal length dan petal width
plt.scatter(petal_length, petal_width)
plt.xlabel('petal Length (cm)') #label untuk sumbu x
plt.ylabel('Sepal Width (cm )') #label untuk sumbu y

plt.title('Scatter Plot dari Sepal Length (cm) dan Sepal Width (cm)')
#digunakan untuk menambahkan judul pada sactterlot dengan mengambil string Scatter Plot dari Sepal Length (cm) dan Sepal Width (cm)' sebagai argumen
plt.show #untuk memanggil atau menampilkan scatter plot yang sudah kita buat
```
